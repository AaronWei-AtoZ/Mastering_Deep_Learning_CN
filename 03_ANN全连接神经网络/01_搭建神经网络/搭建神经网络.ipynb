```python
# 导包
import torch
from torchsummary import summary    # 计算模型参数，查看模型结构
import torch.nn as nn
```

# 神经网络搭建流程

## 1. 定义一个类，继承自nn.Module
## 2. 在__init__方法中，搭建神经网络
## 3. 在forward方法中，完成前向传播。

```python
class MyModel(nn.Module):
    # 1. 在__init__魔法方法中，完成初始化：父类成员和搭建神经网络
    def __init__(self):
        # 1.1 初始化父类成员
        super().__init__()
        # 1.2 搭建神经网络的隐藏层与输出层
        # 隐藏层1：输入特征数3，输出特征数3
        self.linear1 = nn.Linear(3, 3)
        # 隐藏层2：输入特征数3，输出特征数2
        self.linear2 = nn.Linear(3, 2)
        # 输出层：输入特征数2，输出特征数2
        self.output = nn.Linear(2, 2)

        # 1.3 对隐藏层1进行参数初始化
        nn.init.xavier_normal_(self.linear1.weight)
        nn.init.zeros_(self.linear1.bias)

        # 1.4 对隐藏层2进行参数初始化
        nn.init.kaiming_normal_(self.linear2.weight)
        nn.init.zeros_(self.linear2.bias)

    # 2. 前向传播：输入层 -> 隐藏层 -> 输出层
    def forward(self, x):
        # 1.1 第一层 隐藏层计算，加权求和 + 激活函数(sigmoid)
        x = torch.sigmoid(self.linear1(x))

        # 1.2 第二层 隐藏层计算，加权求和 + 激活函数(relu)
        x = torch.relu(self.linear2(x))

        # 1.3 输出层计算：加权求和 + 激活函数(softmax)
        x = torch.softmax(self.output(x), dim=-1)   # dim=-1 -> 按行计算，一条样本一条样本的处理

        # 1.4 返回预测值
        return x

```

## 模型训练

```python
def train():
    model = MyModel()

    # 创建数据集
    data = torch.randn(5, 3)
    print(data)
    print(data.shape)
    print(data.requires_grad)

    # 进行顺联
    out = model(data)
    print(out)
    print(out.shape)
    print(out.requires_grad)


    # 计算 和 查看模型参数
    print('============================ 计算模型参数 ============================')
    summary(model, input_size=(3,))


```

```python
if __name__ == '__main__':
    train()
```
